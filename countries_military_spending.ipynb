```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('SIPRI Military Expenditure Database.xlsx')
```

```python
df
```

```python
df.isna().sum()
```

```python
df=df[(df['Year']>=2000) & (df['Year']<=2020)]
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

```python
spendingpc=df.groupby(['Country'])['Spending per Capita'].mean()
```

```python
spendingpc
```

```python
spendingpc.plot(kind='bar', title = 'Mean Military spending per Capita 2000-2020 (USD)', figsize = (20,10), cmap='Dark2', rot = 90)
```
